# Twitch 게임 스트리밍 인기도 변동

```python
# Load the datasets
twitch_game_data_path = '/Users/USER/Desktop/실습/Twitch_game_data.csv'
twitch_global_data_path = '/Users/USER/Desktop/실습/Twitch_global_data.csv'
twitch_user_data_path = '/Users/USER/Desktop/실습/Twitch_user_data.csv'
dataset_path = '/Users/USER/Desktop/실습/dataset.csv'

twitch_game_df = pd.read_csv(twitch_game_data_path, encoding='ISO-8859-1')
twitch_global_df = pd.read_csv(twitch_global_data_path, encoding='ISO-8859-1')
twitch_user_df = pd.read_csv(twitch_user_data_path, encoding='ISO-8859-1')
dataset_df = pd.read_csv(dataset_path, encoding='ISO-8859-1')

# Display basic information about the datasets
print("Twitch Game Data:")
print(twitch_game_df.head())
print(twitch_game_df.columns)
print(twitch_game_df.info())

print("\nTwitch Global Data:")
print(twitch_global_df.head())
print(twitch_global_df.columns)
print(twitch_global_df.info())

print("\nTwitch User Data:")
print(twitch_user_df.head())
print(twitch_user_df.columns)
print(twitch_user_df.info())

print("\nAdditional Dataset:")
print(dataset_df.head())
print(dataset_df.columns)
print(dataset_df.info())

# Check unique values in the Month column
print("\nUnique values in the Month column of Twitch Game Data:")
print(twitch_game_df['Month'].unique())

# Display datasets as dataframes
print("\nTwitch Game DataFrame:")
display(twitch_game_df.head())

print("\nTwitch Global DataFrame:")
display(twitch_global_df.head())

print("\nTwitch User DataFrame:")
display(twitch_user_df.head())

print("\nAdditional DataFrame:")
display(dataset_df.head())
```

- 이 부분에서는 네 개의 CSV 파일을 로드하고, 데이터프레임으로 변환합니다.
- 각 데이터프레임의 첫 몇 개의 행, 열 이름 및 요약 정보를 출력합니다.
- 'Month' 열의 고유값을 출력하여 데이터의 형식을 확인합니다.

결측치 처리

```python
# Convert date columns to datetime if necessary
if 'Date' in dataset_df.columns:
    dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])
if 'Year' in twitch_game_df.columns:
    twitch_game_df['Year'] = pd.to_datetime(twitch_game_df['Year'], format='%Y')
if 'Month' in twitch_game_df.columns:
    twitch_game_df['Month'] = twitch_game_df['Month'].apply(lambda x: f"{x:02d}")  # Ensure month format is two digits
    twitch_game_df['Month'] = pd.to_datetime(twitch_game_df['Month'], format='%m').dt.month

# Extract year and month for easier analysis
twitch_game_df['Year'] = twitch_game_df['Year'].dt.year
twitch_game_df['Month'] = twitch_game_df['Month']

# Check for any missing values and handle them if necessary
print(twitch_game_df.isnull().sum())
print(twitch_global_df.isnull().sum())
print(twitch_user_df.isnull().sum())
print(dataset_df.isnull().sum())

# Fill or drop missing values
twitch_game_df.fillna(0, inplace=True)
twitch_global_df.fillna(0, inplace=True)
twitch_user_df.fillna(0, inplace=True)
dataset_df.fillna(0, inplace=True)
```

스트리머 수, 평균 시청자 수, 최고 시청자 수를 이용하여 각각 연도 별로 어떤 변화가 있는지 확인 

```python
# Group by year and month, then calculate the mean for the relevant columns
grouped_by_year_month = twitch_game_df.groupby(['Year', 'Month'])[['Streamers', 'Avg_viewers', 'Peak_viewers']].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=grouped_by_year_month, x='Year', y='Streamers', label='Streamers', errorbar=None)
sns.lineplot(data=grouped_by_year_month, x='Year', y='Avg_viewers', label='Avg_viewers', errorbar=None)
sns.lineplot(data=grouped_by_year_month, x='Year', y='Peak_viewers', label='Peak_viewers', errorbar=None)
plt.title('Yearly Comparison of Streamers, Avg_viewers, and Peak_viewers')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()
```

가장 많은 방송을 한 게임 목록 top 10

```python
# Group by game and year, then calculate the mean for the relevant columns
grouped_by_game_year = twitch_game_df.groupby(['Game', 'Year'])[['Streamers', 'Hours_watched', 'Hours_streamed', 'Avg_viewers']].mean().reset_index()

# Select top games based on popularity
top_games = twitch_game_df.groupby('Game')['Streamers'].sum().sort_values(ascending=False).head(10).index
top_games_data = grouped_by_game_year[grouped_by_game_year['Game'].isin(top_games)]

plt.figure(figsize=(14, 7))
for game in top_games:
    game_data = top_games_data[top_games_data['Game'] == game]
    sns.lineplot(data=game_data, x='Year', y='Streamers', label=game, errorbar=None)
plt.title('Yearly Change in Streamers for Top Games')
plt.xlabel('Year')
plt.ylabel('Streamers')
plt.legend()
plt.show()
```

인기 많은 게임 top5

```python
# Calculate the overall popularity of each game
twitch_game_df['Popularity'] = twitch_game_df[['Streamers', 'Avg_viewers', 'Peak_viewers', 'Hours_watched', 'Hours_streamed']].sum(axis=1)
top_5_games = twitch_game_df.groupby('Game')['Popularity'].sum().sort_values(ascending=False).head(5).index
top_5_games_data = twitch_game_df[twitch_game_df['Game'].isin(top_5_games)]

plt.figure(figsize=(14, 7))
for game in top_5_games:
    game_data = top_5_games_data[top_5_games_data['Game'] == game]
    sns.lineplot(data=game_data, x='Year', y='Popularity', label=game, errorbar=None)
plt.title('Yearly Popularity of Top 5 Games')
plt.xlabel('Year')
plt.ylabel('Popularity Score')
plt.legend()
plt.show()
```

# 가설 3: 상위 100 이하의 게임들 중 1년안에 얼마나 순위권으로 들어갈까

```python
# 데이터 로드 및 전처리
twitch_game_df = pd.read_csv('/Users/USER/Desktop/실습/Twitch_game_data.csv', encoding='ISO-8859-1')
if 'Year' in twitch_game_df.columns:
    twitch_game_df['Year'] = pd.to_datetime(twitch_game_df['Year'], format='%Y')
if 'Month' in twitch_game_df.columns:
    twitch_game_df['Month'] = twitch_game_df['Month'].apply(lambda x: f"{x:02d}")
    twitch_game_df['Month'] = pd.to_datetime(twitch_game_df['Month'], format='%m').dt.month
twitch_game_df['Year'] = twitch_game_df['Year'].dt.year
twitch_game_df.fillna(0, inplace=True)

# 순위 계산
twitch_game_df['Rank'] = twitch_game_df.groupby('Year')['Streamers'].rank(ascending=False, method='min')
low_ranked_games = twitch_game_df[twitch_game_df['Rank'] > 100].copy()
low_ranked_games['Next_Year'] = low_ranked_games['Year'] + 1

# 다음 해 데이터와 병합
merged = pd.merge(low_ranked_games, twitch_game_df, left_on=['Game', 'Next_Year'], right_on=['Game', 'Year'], suffixes=('', '_next'))
merged['Entered_Top_100'] = merged['Rank_next'] <= 100
enter_top_100_rate = merged.groupby('Year')['Entered_Top_100'].mean() * 100

# 누적 영역 그래프 시각화
plt.figure(figsize=(14, 7))
plt.fill_between(enter_top_100_rate.index, enter_top_100_rate.values, color='skyblue', alpha=0.4)
plt.plot(enter_top_100_rate.index, enter_top_100_rate.values, marker='o', linestyle='-', color='blue', linewidth=1.5)
plt.title('Percentage of Games Entering Top 100 Within One Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()

# 피크 채널 변화 계산 및 시각화
merged['Peak_channels_change'] = merged['Peak_channels_next'] - merged['Peak_channels']
peak_channels_change_avg = merged.groupby('Year')['Peak_channels_change'].mean()

plt.figure(figsize=(14, 7))
sns.barplot(x=peak_channels_change_avg.index, y=peak_channels_change_avg.values, palette='coolwarm')
plt.title('Average Change in Peak Channels for Games Falling Out of Top 100')
plt.xlabel('Year')
plt.ylabel('Average Change in Peak Channels')
plt.grid(True)
plt.show()
```

```python
# Convert date columns to datetime if necessary
if 'Date' in dataset_df.columns:
    dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])
if 'Year' in twitch_game_df.columns:
    twitch_game_df['Year'] = pd.to_datetime(twitch_game_df['Year'], format='%Y')
if 'Month' in twitch_game_df.columns:
    twitch_game_df['Month'] = twitch_game_df['Month'].apply(lambda x: f"{x:02d}")  # Ensure month format is two digits
    twitch_game_df['Month'] = pd.to_datetime(twitch_game_df['Month'], format='%m').dt.month

# Extract year and month for easier analysis
twitch_game_df['Year'] = twitch_game_df['Year'].dt.year
twitch_game_df['Month'] = twitch_game_df['Month']

# Check for any missing values and handle them if necessary
twitch_game_df.fillna(0, inplace=True)
twitch_global_df.fillna(0, inplace=True)
twitch_user_df.fillna(0, inplace=True)
dataset_df.fillna(0, inplace=True)

# Rank games by Streamers for each year
twitch_game_df['Rank'] = twitch_game_df.groupby('Year')['Streamers'].rank(ascending=False, method='min')

# Filter games ranked 100 or below
low_ranked_games = twitch_game_df[twitch_game_df['Rank'] > 100].copy()

# Shift the year to get the next year
low_ranked_games['Next_Year'] = low_ranked_games['Year'] + 1

# Merge with the next year data to see if the game is still present
merged = pd.merge(low_ranked_games, twitch_game_df, left_on=['Game', 'Next_Year'], right_on=['Game', 'Year'], suffixes=('', '_next'))

# Determine if the game entered the top 100 in the next year
merged['Entered_Top_100'] = merged['Rank_next'] <= 100

# Calculate the percentage of games that entered the top 100 in the next year
enter_top_100_rate = merged.groupby('Year')['Entered_Top_100'].mean() * 100

# Plotting the percentage of games entering the top 100 within one year as a line plot
plt.figure(figsize=(14, 7))
plt.plot(enter_top_100_rate.index, enter_top_100_rate.values, marker='o', linestyle='-', color='coral', linewidth=1.5)
plt.title('Percentage of Games Entering Top 100 Within One Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()

# Calculate the change in Peak_channels
merged['Peak_channels_change'] = merged['Peak_channels_next'] - merged['Peak_channels']

# Calculate the average change in Peak_channels for each year
peak_channels_change_avg = merged.groupby('Year')['Peak_channels_change'].mean()

# Plotting the average change in Peak_channels
plt.figure(figsize=(14, 7))
plt.plot(peak_channels_change_avg.index, peak_channels_change_avg.values, marker='o', linestyle='-', color='blue', linewidth=1.5)
plt.title('Average Change in Peak Channels for Games Falling Out of Top 100')
plt.xlabel('Year')
plt.ylabel('Average Change in Peak Channels')
plt.grid(True)
plt.show()
```
