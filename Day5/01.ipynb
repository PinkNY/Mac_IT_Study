### ì˜ˆì œ: íŒ©í† ë¦¬ì–¼ í•¨ìˆ˜

íŒ©í† ë¦¬ì–¼ í•¨ìˆ˜ëŠ” ì–‘ì˜ ì •ìˆ˜ `n`ì— ëŒ€í•´ 1ë¶€í„° `n`ê¹Œì§€ì˜ ì •ìˆ˜ë¥¼ ëª¨ë‘ ê³±í•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. íŒ©í† ë¦¬ì–¼ì„ ì¬ê·€ì ìœ¼ë¡œ ì •ì˜í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

- `0! = 1` (ê¸°ë³¸ ì¡°ê±´)
- `n! = n * (n-1)!` (ì¬ê·€ ì¡°ê±´)

```python
def factorial(n):
    if n == 0:  # ê¸°ë³¸ ì¡°ê±´
        return 1
    else:
        return n * factorial(n - 1)  # ì¬ê·€ ì¡°ê±´

print(factorial(5))  # ì¶œë ¥: 120
```

ì´ë¥¼ íŒŒì´ì¬ìœ¼ë¡œ êµ¬í˜„í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

```python
def fibonacci(n):
    if n == 0:  # ê¸°ë³¸ ì¡°ê±´
        return 0
    elif n == 1:  # ê¸°ë³¸ ì¡°ê±´
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)  # ì¬ê·€ ì¡°ê±´

print(fibonacci(6))  # ì¶œë ¥: 8
```

### 1. ìˆœì—´ (Permutation)

ì£¼ì–´ì§„ ë¦¬ìŠ¤íŠ¸ì˜ ëª¨ë“  ìˆœì„œë¥¼ íƒìƒ‰í•˜ëŠ” ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•œ ë°©ë²•ì…ë‹ˆë‹¤.

```python
# ì¬ê·€
def permute(data, depth, n, result):
    if depth == n:
        print(result)
        return

    for i in range(len(data)):
        if data[i] not in result:
            result.append(data[i])
            permute(data, depth + 1, n, result)
            result.pop()

data = [1, 2, 3]
permute(data, 0, len(data), [])
```

```python
# ìŠ¤íƒ
def permute(data):
    result = []
    stack = [([], data)]

    while stack:
        perm, items = stack.pop()
        if not items:
            result.append(perm)
        else:
            for i in range(len(items)):
                new_perm = perm + [items[i]]
                new_items = items[:i] + items[i+1:]
                stack.append((new_perm, new_items))

    return result

data = [1, 2, 3]
permutations = permute(data)
for perm in permutations:
    print(perm)

```

### 2. ì¡°í•© (Combination)

ì£¼ì–´ì§„ ë¦¬ìŠ¤íŠ¸ì—ì„œ íŠ¹ì • ê°œìˆ˜ì˜ ìš”ì†Œë¥¼ ì„ íƒí•˜ëŠ” ëª¨ë“  ê²½ìš°ë¥¼ ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•˜ì—¬ íƒìƒ‰í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

```python
# ì¬ê·€
def combine(data, start, n, k, result):
    if len(result) == k:
        print(result)
        return

    for i in range(start, n):
        result.append(data[i])
        combine(data, i + 1, n, k, result)
        result.pop()

data = [1, 2, 3]
k = 2  # ì¡°í•©ì˜ í¬ê¸°
combine(data, 0, len(data), k, [])
```

```python
# ìŠ¤íƒ
def combine(data, k):
    result = []
    stack = [(0, [])]

    while stack:
        start, combination = stack.pop()
        if len(combination) == k:
            result.append(combination)
        else:
            for i in range(start, len(data)):
                new_combination = combination + [data[i]]
                stack.append((i + 1, new_combination))

    return result

data = [1, 2, 3]
k = 2
combinations = combine(data, k)
for comb in combinations:
    print(comb)

```

### 3. ë¶€ë¶„ ì§‘í•© (Subset)

ì£¼ì–´ì§„ ë¦¬ìŠ¤íŠ¸ì˜ ëª¨ë“  ë¶€ë¶„ ì§‘í•©ì„ ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•˜ì—¬ íƒìƒ‰í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

```python
# ì¬ê·€
def subset(data, index, result):
    if index == len(data):
        print(result)
        return

    subset(data, index + 1, result)
    result.append(data[index])
    subset(data, index + 1, result)
    result.pop()

data = [1, 2, 3]
subset(data, 0, [])
```

```python
# ìŠ¤íƒ
def subset(data):
    result = [[]]

    for item in data:
        new_subsets = []
        for subset in result:
            new_subset = subset + [item]
            new_subsets.append(new_subset)
        result.extend(new_subsets)

    return result

data = [1, 2, 3]
subsets = subset(data)
for s in subsets:
    print(s)

```

```python
def permute(s):
    result = []
    stack = [('', s)]

    while stack:
        perm, chars = stack.pop()
        if not chars:
            result.append(perm)
        else:
            for i in range(len(chars)):
                stack.append((perm + chars[i], chars[:i] + chars[i+1:]))
    return result

s = "ABC"
permutations = permute(s)
for perm in permutations:
    print(perm)
```
