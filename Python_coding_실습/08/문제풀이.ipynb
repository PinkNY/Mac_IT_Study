```python
import heapq

def solution(array, commands):
    answer = []
    
    for command in commands:
    # commands 리스트를 순회하며 각 명령을 command에 할당합니다.
        start, end, k = command
        sub_array = array[start-1:end]
        # array에서 start-1부터 end까지의 부분 배열을 sub_array에 할당합니다.
        
        heapq.heapify(sub_array)
        # heapq.heapify(sub_array)를 사용하여 부분 배열을 힙으로 변환합니다
        smallest = heapq.nsmallest(k, sub_array)[-1]
        # sub_array 에서 k 번째로 작은 원소를 찾는다.
        # 이 함수는 k 개의 가장 작은 요소들을 리스트로 반환,
        # k번째 요소를 얻기 위해 [-1] 인덱싱을 사용.
        
        answer.append(smallest)
        
    return answer
```

```python
def solution(nubers):
    numbers = list(map(str, numbers))
    
    for i in range(1, len(nubers)):
        key = numbers[i]
        j = i -1
        
        while j >= 0 and (numbers[j] + key < key + numbers[j]):
            numbers[j + 1] = nubers[j]
            j -= 1
        nubers[j + 1] = key
        
    return str(int(''.join(numbers)))
```

```python
def solution(citations):
    n = len(citations)
    
    count = [0] * (n +1)
    
    for c in citations:
        if c >= n:
            count[n] += 1
        else:
            count[c] += 1
            
    total = 0
    
    for i in range(n, -1, -1):
        total += count[i]
        if total >= i:
            return i
        
    return 0
```

```python
def solution(citations):
    citations.sort()
    n = len(citations)
    left, right = 0, n
    
    while left < right:
        mid = (left + right) // 2
        if citations[mid] >= n - mid:
            right = mid
        else:
            left = mid + 1
            
    return n - left
```

```python
from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        merged = []
        for i in sorted(intervals, key=lambda x: x[0]):
            if merged and i[0] <= merged[-1][1]:
                merged[-1][1] = max(merged[-1][1], i[1])
            else:
                merged += i,
                # merged += i, = 튜플로 생성 한다. 콤마가 튜플을 생성하는 역할.
                # 콤마는 중첩 리스트로 만들어주는 역할을 한다.
                # 대괄호 사용하는것과 같아서 편한 방법으로 하면 됨.
        return merged
```

```python
from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        result = []
        for i in sorted(intervals):
            if result and i[0] <= result[-1][1]:
                result[-1][1] = max(result[-1][1], i[1])
            else:
                result += [i]
        return result
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        cur = parent = ListNode(0)
        while head:
            while cur.next and cur.next.val < head.val:
                cur = cur.next
            
            cur.next, head.next, head = head, cur.next, head.next
            
            if head and cur.val > head.val:
                cur = parent
                
        return parent.next
```

```python
from typing import List
import heapq

class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        heap = []
        
        for (x,y) in points:
            dist = x ** 2 + y ** 2
            # dist = math.sqrt(x ** 2 + y ** 2)
            # import math 하면 재곱근 계산 가능.
            heapq.heappush(heap, (dist, x, y))
            
        result = []
        
        for _ in range(k):
            _, (x, y) = heapq.heappop(heap)
            result.append((x,y))
            
        return result
```

```python
import sys

class Solution:
    def reverse(self, x: int) -> int: # 입력 받은 숫자를 문자열로 변환
        strx = str(x) 
        
        if x >= 0:  # 숫자가 0 이상일 경우 아래 슬라이스로 뒤집는다.
            res = strx[::-1]
        else:   # 음수일 경우 앞의 마이너스 기호를 때고 뒤집는다.
            temp = strx[1:]
            temp2 = temp[::-1]
            res = '-' + temp2
            
        if int(res) >= 2**31-1 or int(res) <= -2**31:
            return 0    # 출력할 답이 정수 32비트를 넘을 경우 0을 반환한다.
        else:   # 아니면 정수로 치환하여 반환한다.
            return int(res)
```
