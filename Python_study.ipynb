```python
import string
import unittest

# 비밀번호가 충분히 안전한지 확인합니다.
def check_password(password):
    
    # password가 문자열이 아니면 False
    if type(password) is not str:
        return False

    # 비밀번호가 8자리보다 짧으면 False
    if password.length < 8:  # 'length'를 'len'으로 변경
        return False
    
    # 비밀번호가 알파벳, 숫자, 특수문자를 하나라도 포함하지 않으면 False
    check_alphabet, check_digit, check_symbol = False, False, False
    
    for letter in string.ascii_letters:
        if letter in password:
            check_alphabet = True
    
    for digit in string.digits:
        if digit in password:  # 'string.digits'를 'password'로 변경
            check_digit = True
    
    for symbol in string.punctuation:
        if symbol in password:
            check_symbol = True
    
    # 3개의 변수 중 하나라도 True가 아니라면 False를 return
    if all([check_alphabet, check_digit, check_symbol]):  # 'all' 함수에 리스트를 전달
        return False
    
    for whitespace in string.whitespace:
        if whitespace in password:  # 'not'을 제거
            return False

    return True  # 모든 조건을 만족하면 True를 반환


class CheckPasswordTests(unittest.TestCase):
    def test_letters_only(self):
        self.assertFalse(check_password, "password")
    
    def test_digits_only(self):
        self.assertFalse(check_password, "12345678")
        
    def test_whitespace(self):
        self.assertFalse(check_password, "Hello World 2019!")
    
    def test_strong_password(self):
        self.assertTrue(check_password, "Happy-New-Year-2019")



```

```python
import string
import unittest

# 비밀번호가 충분히 안전한지 확인합니다.
def check_password(password):
    
    # password가 문자열이 아니면 False
    if type(password) is not str:
        return False

    # 비밀번호가 8자리보다 짧으면 False
    if len(password) < 8:
        return False
    
    # 비밀번호가 알파벳, 숫자, 특수문자를 하나라도 포함하지 않으면 False
    check_alphabet, check_digit, check_symbol = False, False, False
    
    for letter in string.ascii_letters:
        if letter in password:
            check_alphabet = True
    
    for digit in string.digits:
        if digit in password:
            check_digit = True
    
    for symbol in string.punctuation:
        if symbol in password:
            check_symbol = True
    
    # 3개의 변수 중 하나라도 True가 아니라면 False를 return
    if not all([check_alphabet, check_digit, check_symbol]):
        return False
    
    for whitespace in string.whitespace:
        if whitespace in password:
            return False

    return True


class CheckPasswordTests(unittest.TestCase):
    def test_letters_only(self):
        self.assertFalse(check_password("password"))
    
    def test_digits_only(self):
        self.assertFalse(check_password("12345678"))
        
    def test_whitespace(self):
        self.assertFalse(check_password("Hello World 2019!"))
    
    def test_strong_password(self):
        self.assertTrue(check_password("Happy-New-Year-2019"))


if __name__ == "__main__" :
    unittest.main()
```

```python
# 사용자로부터 숫자를 입력받습니다.
numbers = input("숫자를 입력하세요 (쉼표로 구분): ")

# 입력받은 숫자를 쉼표로 구분하여 리스트로 변환합니다.
numbers = list(map(int, numbers.split(',')))

# 홀수와 짝수를 저장할 빈 리스트를 생성합니다.
odd_numbers = []
even_numbers = []

# 리스트의 각 숫자를 확인하여 홀수와 짝수를 분류합니다.
for number in numbers:
    if number % 2 == 0:
        even_numbers.append(number)
    else:
        odd_numbers.append(number)

# 홀수와 짝수 리스트를 출력합니다.
print("홀수 리스트: ", odd_numbers)
print("짝수 리스트: ", even_numbers)
```
